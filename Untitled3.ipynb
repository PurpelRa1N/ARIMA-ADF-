```python
from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(df['Milk in Pounds per Cow'])
```

```python
def adf_check(time_series):
  result = adfuller(time_series)
  print("Augmented Dicky-Fuller Test")
  labels = ['ADF Test Statistic', 'p-value', '# of lags', 'Num of Observations used']

  for value, label in zip(result, labels):
    print(label+ " : " + str(value))

  if result[1] <= 0.05:
    print("strong evidence against null hypothesis")
    print("reject null hypothesis")
    print("Data has no unit root and is stationary")

  else:
    print("weak evidence against null hypothesis")
    print("Failed to reject null hypothesis")
    print("Data has a unit root, it is non stationary")
    
```

```python
adf_check(df['Milk in Pounds per Cow'])
```

```python
df['First Differnece'] = df['Milk in Pounds per Cow'] - df['Milk in Pounds per Cow'].shift(1)
```

```python
df['First Differnece'].plot()
```

```python
adf_check(df['First Differnece'].dropna())
```

```python
df['Milk Second Difference'] = df['First Differnece'] - df['First Differnece'].shift(1)
```

```python
adf_check(df['Milk Second Difference'].dropna())
```

```python
df['Milk Second Difference'].plot()
```

```python
df['Seasonal Differnece'] = df['Milk in Pounds per Cow'] - df['Milk in Pounds per Cow'].shift(12)
```

```python
adf_check(df['Seasonal Differnece'].dropna())
```

```python
df['Seasonal Differnece'].plot()
```
